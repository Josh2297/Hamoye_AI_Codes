```python
# What is the total number and percentage of missing data in 2016 to 2 decimal places?
missing_data=food_data['Y2016'].isnull().sum()
print('Number of Missing Data in Y2016: ',missing_data)
print()
percentage=(missing_data/len(food_data)*100)
print('The Percentage of Missing Data in Y2016 is {:.2f}:'.format(percentage))
```

```python
# What is the total Import Quantity in Algeria in 2018?
ques19=food_data[food_data.Area.isin(['Algeria'])] # Selecting a Subset of Algeria Only
ques19
```
